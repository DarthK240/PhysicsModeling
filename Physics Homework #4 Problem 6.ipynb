```python
# Import necessary packages
import matplotlib.pyplot as plt # For visualization
import numpy as np # For ease in calculation
from scipy.optimize import curve_fit # Using to fit the curve to find the best possible value of A
```

```python
# Defining the Semi-major Axes of the eight major planets' orbits (in AU)
semi_major_axes = [0.39, 0.72, 1.00, 1.52, 5.20, 9.58, 19.22, 30.05]

# Defining the periods of orbit of the eight major planets (in years)
orbital_periods = [0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.79]

# Defining the Kepler's law equation using a function
def keplers_law(x, A):
    return A * x**1.5
```

```python
# Now, we can fit the data from the lists initialized above
# to the Kepler's law equation, finding the optimal value of A
optimized_parameters, _ = curve_fit(keplers_law, semi_major_axes, orbital_periods)

# Extracting the best possible value of A
A_optimal = optimized_parameters[0]
```

```python
# We can now start to create the log-log plot of P vs. a for the planets, represented by red dots
plt.figure(figsize=(10, 6))
plt.loglog(semi_major_axes, orbital_periods, 'ro', label='Planets')

# Plotting the calculated values of y = Ax^3/2 using the newly optimized A
x_range = np.linspace(0.2, 35, 100)
y_optimal = A_optimal * x_range**1.5
plt.loglog(x_range, y_optimal, 'b--', label='y = Ax^3/2 (Optimized)')

# Labeling the axes of the graph and adding a legend
plt.xlabel('Semi-Major Axis (AU)')
plt.ylabel('Orbital Period (Years)')
plt.legend()

# Final Step: Graph visualization and printing the optimal value of A
plt.grid(True)
plt.title('Kepler\'s Law: P vs. a')
plt.show()

print("Optimal value of A:", A_optimal)
```
